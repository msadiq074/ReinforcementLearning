```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

### Environment For A MultiArm Bandit Scenario

```python

class MultiArmBanditEnv:
    def __init__(self,n):
        
        # n = number of arms of the multiarm bandit scenario
        self.arms = n
        self.rewards = np.random.uniform(low=0, high=100, size=(self.arms,))
    
    def get_reward(self,index):
        # Method returns the Reward for selecting a particular arm
        
        # Adding Noise to the Reward Returned. Noise is Based on Gaussian Distribution with mean 0,
        # variance 1 and length of array as 1
        noise = np.random.normal(0,1,1)
        reward = self.rewards[index]
        
        # Adding Noise
        reward = reward + noise
        return reward
    
    def get_actual_reward(self):
        
        # Debug Function. Used to see if Function estimates the rewards close to the real ones
        return self.rewards
    
    def get_arms(self):
        
        # Returns the number of Arms of the env setup
        return self.arms
```

### Agent Used for this scenario

```python

class Agent:
    def __init__(self,arms,epsilon):
        
        # arms = number of arms in the multiarm bandit
        # epsilon = Probability of dropping the ArgMax(q(a)) and going with a lesser reward arm. (To promote exploration)
        self.arms = arms
        self.epsilon = epsilon
        
        # Record of all the rewards.
        self.Q_t = list(np.random.uniform(low=0, high=50, size=(self.arms,)))
        
        # Record of number of times the arm has been chosen
        self.N_t = [0]*arms
        
        # Record of cummulative rewards gained at each turn
        self.record = [0]
        
        # Record of average cummulative rewards at each turn
        self.avg_reward = [0]
        
    
    def select_arm(self):
        # Method returns the index of the arm to be chosen
        
        # implementation of dropping an arm by generating a number at random
        num = random.random()
        if num<=(1-self.epsilon):
            
            # The argmax(q(a)) arm has been chosen hence
            return self.Q_t.index(max(self.Q_t))
        else:
            # The argmax(q(a)) arm wasn't chosen
            
            ''' 
            Here the rest of the arms have equal probability to be selected hence the random number generator
            is used to figure out which arm to select
            '''
            
            max_index = self.Q_t.index(max(self.Q_t))
            new_num = random.random()
            index = int(new_num/(1/(self.arms-1)))
            
            # If selected index is to the right of the max_reward arm
            if index>=max_index:
                return index+1
            else:
                return index
            
    def update(self,index,reward):
        # Method updates the Q_t, N_t array of the agent
        
        # Add the reward achieved
        self.record.append( self.record[len(self.record)-1] + reward )
        self.avg_reward.append(self.record[len(self.record)-1]/(len(self.avg_reward)))
        
        # Update the Q_t array.
        try:
            self.Q_t[index] = self.Q_t[index] + (1/self.N_t[index]) * (reward - self.Q_t[index])
        except:
            self.Q_t[index] = self.Q_t[index] + (reward - self.Q_t[index])
        
        # Update the N_t array
        self.N_t[index] +=1
        
    def plot(self):
        # plots the cummulative reward achieved by the agent
        plt.plot(self.record)
        plt.show()
            
    def plot_avg(self):
        # plots the average reward achieved by the agent
        plt.plot(self.avg_reward)
        plt.show()
    
    def get_avg(self):
        # return the record of average reward achieved by the agent
        return self.avg_reward
    
    def get_Q(self):
        #returns the Q_t array. Used for debugging to see the estimate of the rewards 
        return self.Q_t

    def get_N(self):
        # To check how many times a particular arm was chosen
        return self.N_t
```

### Simulating the Agent Environment interaction

```python
 
env = MultiArmBanditEnv(10)

agent = Agent(env.get_arms(),0.1)
agent2 = Agent(env.get_arms(),0)

for i in range(2000):
    
    # Selection of the arm by the agent
    arm = agent.select_arm()
    
    # Reward gotten from the environment for the specific arm
    reward = env.get_reward(arm)
    
    # Updation of the particular selection of the arm through the reward
    agent.update(arm,reward)

    # Same steps for Agent 2
    arm2 = agent2.select_arm()
    reward2 = env.get_reward(arm2)
    agent2.update(arm2,reward2)

    
reward1= agent.get_avg()
reward2 = agent2.get_avg()

# Plotting the average rewards from different Agents
plt.plot(reward1,label="E=0.1")
plt.plot(reward2,label="E=0")
plt.legend()
plt.show()
```
